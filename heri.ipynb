```python
import geopandas as gpd #mengimport modul geopandas dengan memberi inisial gpd
```

```python
gdf = gpd.read_file("RBI25K_ADMINISTRASI_LN.json") #membaca file geospasial format json
gdf.head() #menampilkan lima baris pertama dari dataframe gdf
```

```python
gdf = gdf[['NAMOBJ', 'FCODE', 'REMARK', 'METADATA', 'geometry']]  #memilih kolom dari dataframe gdf
```

```python
pip install folium matplotlib mapclassify #menginstall folium matplotllib dan mapclassify
```

```python
import folium #mengimport folium yang digunakan membuat peta interaktif
m = gdf.explore(column="METADATA", name="Meta Data") #membuat peta interaktif dengan menggunakan data dari gdf pada kolom metadata dengan nama meta data
folium.LayerControl().add_to(m) #menambahkan kontrol layer ke peta interaktif m
m #menampilkan peta interaktif
```
